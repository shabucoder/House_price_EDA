```python
### this cell gives tthe features with Nan Values

featuresWith_Nan=[ i for i in df.columns if df[i].isnull().sum()>=1]
len(featuresWith_Nan)
```

```python
for i in featuresWith_Nan:
    print(i,'=',np.round(df[i].isnull().mean()*100,4), '% missing Values')
```

```python
sns.boxplot(df['SalePrice'])
```

```python
data =df.copy()
for i in featuresWith_Nan:
  data[i]=np.where(df[i].isnull(),1,0) ## this line changes the values if its nan values then changes to 1, else it will be 0
  data.groupby(i)['SalePrice'].median().plot.bar()
  plt.title(i)
  plt.show()
```
